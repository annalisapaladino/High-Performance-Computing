### Processes vs Latency 

```python
# Processes vs Latency BCAST 1

fig, ax = plt.subplots()
colors = plt.cm.viridis(np.linspace(0, 1, len(bcast0_core_thin['Size'].unique())))
for i, c in zip(bcast0_core_thin['Size'].unique(), colors):
    ax.plot(bcast0_core_thin[bcast0_core_thin['Size'] == i]['Processes'], bcast0_core_thin[bcast0_core_thin['Size'] == i]['Latency'], label='Size = %d' % i, color=c)
ax.set_xlabel('Processes')
ax.set_ylabel('Latency')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.axvline(x=12, color='gray', linestyle='--')
plt.axvline(x=24, color='gray', linestyle='--')
plt.ylim(0, 2000)
plt.title('bcast 0 - core')
# plt.savefig('plots/b1_epyc_map_by_core_color_legend.png', bbox_inches='tight')
plt.show()
```
