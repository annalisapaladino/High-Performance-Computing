```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# SCATTER

Read all the csv

```python
# Read scatter0: core, node & socket
scatter0_core_thin = pd.read_csv('scatter0_core_thin.csv', header=0)
scatter0_node_thin = pd.read_csv('scatter0_node_thin.csv', header=0)
scatter0_socket_thin = pd.read_csv('scatter0_socket_thin.csv', header=0)
```

```python
# Read scatter1: core, node & socket
scatter1_core_thin = pd.read_csv('scatter1_core_thin.csv', header=0)
scatter1_node_thin = pd.read_csv('scatter1_node_thin.csv', header=0)
scatter1_socket_thin = pd.read_csv('scatter1_socket_thin.csv', header=0)

scatter1_pt2_core_thin = pd.read_csv('scatter1_pt2_core_thin.csv', header=0)
scatter1_pt2_node_thin = pd.read_csv('scatter1_pt2_node_thin.csv', header=0)
scatter1_pt2_socket_thin = pd.read_csv('scatter1_pt2_socket_thin.csv', header=0)

# merging
scatter1_core_thin = pd.concat([scatter1_core_thin, scatter1_pt2_core_thin], ignore_index=True)
scatter1_node_thin = pd.concat([scatter1_node_thin, scatter1_pt2_node_thin], ignore_index=True)
scatter1_socket_thin = pd.concat([scatter1_socket_thin, scatter1_pt2_socket_thin], ignore_index=True)
```

```python
# Read scatter2: core, node & socket
scatter2_core_thin = pd.read_csv('scatter2_core_thin.csv', header=0)
scatter2_node_thin = pd.read_csv('scatter2_node_thin.csv', header=0)
scatter2_socket_thin = pd.read_csv('scatter2_socket_thin.csv', header=0)

scatter2_pt2_core_thin = pd.read_csv('scatter2_pt2_core_thin.csv', header=0)
scatter2_pt2_node_thin = pd.read_csv('scatter2_pt2_node_thin.csv', header=0)
scatter2_pt2_socket_thin = pd.read_csv('scatter2_pt2_socket_thin.csv', header=0)

# merging
scatter2_core_thin = pd.concat([scatter2_core_thin, scatter2_pt2_core_thin], ignore_index=True)
scatter2_node_thin = pd.concat([scatter2_node_thin, scatter2_pt2_node_thin], ignore_index=True)
scatter2_socket_thin = pd.concat([scatter2_socket_thin, scatter2_pt2_socket_thin], ignore_index=True)
```

### Processes vs Latency 

```python
# Processes vs Latency SCATTER 0
fig, ax = plt.subplots()
colors = plt.cm.viridis(np.linspace(0, 1, len(scatter0_core_thin['Size'].unique())))
for i, c in zip(scatter0_core_thin['Size'].unique(), colors):
    ax.plot(scatter0_core_thin[scatter0_core_thin['Size'] == i]['Processes'], scatter0_core_thin[scatter0_core_thin['Size'] == i]['Latency'], label='Size = %d' % i, color=c)
ax.set_xlabel('Processes')
ax.set_ylabel('Latency')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.axvline(x=12, color='gray', linestyle='--')
plt.axvline(x=24, color='gray', linestyle='--')
plt.ylim(0, 9000)
plt.title('scatter 0 - core')
# plt.savefig('plots/b1_epyc_map_by_core_color_legend.png', bbox_inches='tight')
plt.show()
```
